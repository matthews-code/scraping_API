### Get articles via newsapi

```python
from newsapi import NewsApiClient
from datetime import datetime
from datetime import timezone
import pandas as pd

pd.set_option('display.max_colwidth', None)

# Initialize client
newsapi = NewsApiClient(api_key='f563f261c4f741a28f330d7db0a312b5')


all_articles = newsapi.get_everything(sources='techcrunch',    # More examples: 'bbc-news', 'the-verge', 'abc-news'
                                      from_param='2022-05-11', # from date
                                      to='2022-05-12',         # to date
                                      language='en',           # language
                                      sort_by='popularity',    # sort by
                                      )

newsArticles = all_articles.get('articles')
```

### Clean and format the articles returned then collect the needed information

```python
masterList = []

for currentArticle in newsArticles:
    keysToDelete = ['source', 'url', 'urlToImage', 'description']
    for key in keysToDelete:
        del currentArticle[key]
        
    dateTime = datetime.fromisoformat(currentArticle.get('publishedAt')[:-1]).astimezone(timezone.utc)
    dateTime = dateTime.strftime('%Y-%m-%d')
    
    title = currentArticle.get('title')
    body = currentArticle.get('content')
    author = currentArticle.get('author')
    
    masterList.append((dateTime, title, body, author))
```

### Transfer list into data frame

```python
dFMasterList = pd.DataFrame(masterList)
dFMasterList.columns = ['Date', 'Title', 'Body', 'Author']
```

### Transfer data frame to json file

```python
dFMasterList.to_json('NewsArticles.json', orient='index')
```
